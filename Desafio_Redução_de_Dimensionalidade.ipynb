```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data
import math
from skimage.color import rgb2gray
import skimage

def to_gray(image):

  gray_image = rgb2gray(image) #convert to grayscale

  return gray_image

def to_binary(image):
  gray_image = to_gray(image)
  threshold = skimage.filters.threshold_otsu(gray_image) #input value
  binary_image = (gray_image < threshold) #binary is when less than trheshold

  return binary_image
```

```python
image = data.astronaut()

plt.imshow(image)
```

```python
gray_image = to_gray(image)

plt.imshow(gray_image, cmap='gray')
```

```python
bw_image = to_binary(image)

plt.imshow(bw_image, cmap='gray')
```
